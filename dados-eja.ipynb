É possível perceber os valores ausentes nessas 13 colunas seguem sendo identificadas como object. Portanto, para resolver esse problema irei substituir esses valores ausentes resolvi optar pela técnica de imputação Mediana, pela sua robustêz e por ser pouco afetada pelos outliers que podem sugir se eu usasse a média. 

```python
# Iniciando a substituição de "--" para NaN nas colunas com dados numéricos

col_num = ['EJAFUNG1', 'EJAFUNG2', 'EJAFUNG3', 'EJAFUNG4', 'EJAFUNG5', 'EJAMED1', 'EJAMED2', 'EJAMED3', 'EJAMED4', 'EJAMED5'] #minhas colunas numéricas
```

```python
afd_eja = afd_eja.copy()
```

```python
#Substituindo "--" por nan apenas nessas colunas

afd_eja.loc[:, col_num] = afd_eja[col_num].replace('--', np.nan)
```
