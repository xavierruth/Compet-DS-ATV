### Visualizando os dados - EJA

Dicionário do dataframe afd_eja (adequação da formação docente - EJA)

- GEO: Unidade geográfica
- LOC: Zona rural ou urbana
- DA: Dependência administrativa (estadual, municipal, federal e privado)
- EJAFUNG1 a EJAFUNG5: Educação de Jovens e Adultos no Ensino Fundamental nos grupos 1 a 5
- EJAMED1 a EJAMED5: Educação de Jovens e Adultos no Ensino Médio nos grupos de 1 a 5
- Grupo 1: Representa docentes que possuem a formação adequada
- Grupo 2 a Grupo 5: Representa docentes sem formação adequada em diferentes níveis

```python
# Mediana de docentes com formação adequada por dependência administrativa
plt.figure(figsize=(12, 6))
afd_eja.groupby("DA")[["EJAFUNG1", "EJAMED1"]].median().sort_values("EJAMED1", ascending=False).plot(kind="bar")
plt.title("Percentual de Docentes com Formação Adequada por Dependência Administrativa")
plt.ylabel("Porcentagem")
plt.xlabel("Dependência Administrativa")
plt.legend(["Ensino Fundamental - EJA", "Ensino Médio - EJA"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Comparação entre zona urbana e rural para o EJA no Ensino Fundamental e Ensino Médio
plt.figure(figsize=(10, 6))
afd_eja.groupby("LOC")[["EJAFUNG1", "EJAMED1"]].median().plot(kind="bar")
plt.title("Mediana de Docentes com Formação Adequada por Localização")
plt.ylabel("Porcetagem")
plt.xlabel("Localização")
plt.legend(["Ensino Fundamental - EJA", "Ensino Médio - EJA"])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Top 10 unidades geográficas com maior percentual de formação adequada no Ensino Médio - EJA
plt.figure(figsize=(12, 6))
afd_eja.groupby("GEO")["EJAMED1"].mean().sort_values(ascending=False).head(10).plot(kind="barh")
plt.title("Top 10 Unidades Geográficas com Maior Percentual de Formação Adequada - Ensino Médio")
plt.xlabel("Percentual de docentes com formação adequada")
plt.ylabel("Unidade Geográfica")
plt.tight_layout()
plt.show()
```

```python
# Top 10 unidades geográficas com maior percentual de formação adequada no Ensino Fundamental - EJA
plt.figure(figsize=(12, 6))
afd_eja.groupby("GEO")["EJAFUNG1"].mean().sort_values(ascending=False).head(10).plot(kind="barh")
plt.title("Top 10 Unidades Geográficas com Maior Percentual de Formação Adequada - Ensino Fundamental")
plt.xlabel("Percentual de docentes com formação adequada")
plt.ylabel("Unidade Geográfica")
plt.tight_layout()
plt.show()
```

### Conclusão

De um modo geral, foi possível observar que há uma diferença expressiva entre a formação adequada no Ensino Fundamental do que o Ensino Médio, logo o Ensino Médio EJA supera em quase 67,78%. Além disso há diferenças significativas entre as dependências administrativas, onde a Federal possui maiores índices de adequação no grupo 1, enquanto que nas redes Municipais e Estaduais a maioria dos docentes possuem graduação, mas atuam numa área diferente ou sequer possuem qualquer formação superior completa. 

Também é percebido que dependendo da localização, zona rural e urbana, há uma maior disparidade. Nesse caso a zona urbana concentra mais docentes com formação adequada do que escolas localizadas nas zonas rurais. Logo, a análise desse indicador educacional aponta uma realidade preocupantes presente no dia a dia da EJA, reforçando a necessidade de políticas públicas, capacitação e incentivo a formação docente principalmente em zonas rurais.
